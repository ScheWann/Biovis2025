```python
diffexp = pd.read_csv("../Data/square_008um/analysis/diffexp/gene_expression_kmeans_10_clusters/differential_expression.csv")
diffexp
```

```python
log2fc_threshold = 0  # Log2 fold change threshold
pval_threshold = 0.05 # p value threshold

cluster_columns = [col for col in diffexp.columns if "Cluster" in col]
results = {}

# cluster numbers
for cluster in range(1, 11):
    log2fc_col = f"Cluster {cluster} Log2 fold change"
    pval_col = f"Cluster {cluster} Adjusted p value"

    if all(col in diffexp.columns for col in [log2fc_col, pval_col]):
        # filter
        # filtered_genes = diffexp[
        #     (diffexp[log2fc_col] >= log2fc_threshold) &
        #     (diffexp[pval_col] <= pval_threshold) &
        #     (diffexp[mean_col] > 1)  # This line ensures high mean counts
        # ][["Feature ID", "Feature Name", mean_col, log2fc_col, pval_col]]

        filtered_genes = diffexp[
            (diffexp[log2fc_col] > log2fc_threshold) &
            (diffexp[pval_col] < pval_threshold)
        ][["Feature ID", "Feature Name", log2fc_col, pval_col]]

        # sorted by log2 fold change
        filtered_genes = filtered_genes.sort_values(by=[log2fc_col, pval_col], ascending=False)

        results[f"Cluster {cluster}"] = filtered_genes

# Checking the top 10 genes for each cluster
for cluster, genes in results.items():
    print(f"{cluster}")
    top_genes = genes.head(10)

    log2fc_col = f"{cluster} Log2 fold change"
    pval_col = f"{cluster} Adjusted p value"

    for _, row in top_genes.iterrows():
        # print(row['Feature Name'])
        print(f"Feature Name: {row['Feature Name']}, Log2 Fold Change: {row[log2fc_col]}, Adjusted p value: {row[pval_col]}")
```
