```python
cdata = sc.read_h5ad('../Example_Data/H1-TXK6Z4X-A1_2um/skin_TXK6Z4X_A1_2um_b2c_qc.h5ad')
cdata
```

```python
cdata = cdata[cell_ids].copy()

sc.pp.highly_variable_genes(cdata, n_top_genes=2000, flavor="seurat_v3")

sc.pp.normalize_total(cdata)
sc.pp.log1p(cdata)
sc.pp.scale(cdata, max_value=10)

sc.tl.pca(cdata, use_highly_variable=True)
sc.pp.neighbors(cdata, n_neighbors=15, n_pcs=30)
sc.tl.umap(cdata)

# custom umap for subset of cells
cdata.obsm['X_umap_subset'] = cdata.obsm['X_umap'].copy()

sc.tl.leiden(cdata, resolution=1)
sc.pl.umap(cdata, color=['leiden'], size=50)
```

```python
embedding = cdata.obsm['X_umap_subset'] # shape: (n_cells, 2)
cdata.obs['leiden_subset'] = cdata.obs['leiden'].copy()
cluster_labels = cdata.obs['leiden_subset'].astype(int).values

cell_ids_filtered = cdata.obs_names.tolist()

# Return the result
result = []
for i in range(len(cell_ids_filtered)):
    result.append({
        'id': cell_ids_filtered[i],
        'x': float(embedding[i, 0]),
        'y': float(embedding[i, 1]),
        'cluster': f'Cluster {cluster_labels[i] + 1}'
    })

result
```

```python
cluster_id = 3
sc.tl.rank_genes_groups(cdata, groupby='leiden_subset', method='wilcoxon')
cluster_name = str(cluster_id)
top_genes = cdata.uns['rank_genes_groups']['names'][cluster_name][:100].tolist()
```
